```python
# Importing libraries
import random
import string

# Choosing random word out of present list
def get_random_word(word_list):
    word = random.choice(word_list)
    return word

# Defining function for Hangman game
def hangman():
    word = get_random_word(word_list).upper()
    word_letter = set(word) # letters in the word
    alphabet = set(string.ascii_uppercase)
    used_letter = set() # What the user has used

    lives = 6        
    words = ["-" for _ in word]
    
    # Get user input
    while len(word_letter) > 0 and lives > 0:
    
        # letters used
        print('You have' ,lives,'lives left and You have used these letters',' '.join(used_letter))
        print('current word : ',' '.join(words))
        
        user_letter = input('Guess a letter :').upper()
                
        if len(user_letter) == 1 and user_letter.isalpha():
            if user_letter in alphabet - used_letter:
                used_letter.add(user_letter)
                correct_guess = False # Track if the user's letter is correct
                
                for i in range(len(word)):
                    if word[i] == user_letter:
                        words[i] = user_letter
                        correct_guess = True
                        
                if not correct_guess:
                    lives -= 1
                    print('Letter is not in the word')
                
            elif user_letter in used_letter :
                print(f'You have already used that character. Try again')

        else:
            print('Invalid character. Please entre a single character')
        
    if lives == 0:
        print(f'Sorry you lost! unable to guess the word {word}')
    else:
        print(f'You guess the word {word}')
```
